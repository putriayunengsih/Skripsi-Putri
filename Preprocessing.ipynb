# **FEATURE SELECTION**

```python
features_datajalan = kumpulan_datajalan[["Sub Kategori", "Rusak Ruas Kiri Jalan", "Rusak Ruas Kanan Jalan", "Nama Wilayah","Panjang (M)", "Lebar (M)", "Tebal/Tinggi (M)"]].copy()

# Menampilkan data dalam bentuk tabel
features_datajalan
```

```python
# Print contoh data kosong dalam bentuk tabel
data_preprocessing = features_datajalan.dropna()
data_preprocessing
```

```python
data_preprocessing.isnull().sum()
```

```python
# Mengonversi kolom 'Local_Amount' ke tipe data string
data_preprocessing['Panjang (M)'] = data_preprocessing['Panjang (M)'].astype(str)
data_preprocessing['Lebar (M)'] = data_preprocessing['Lebar (M)'].astype(str)
data_preprocessing['Tebal/Tinggi (M)'] = data_preprocessing['Tebal/Tinggi (M)'].astype(str)

# Menghilangkan koma dari kolom 'Local_Amount'
data_preprocessing['Panjang (M)'] = data_preprocessing['Panjang (M)'].str.replace(',', '').astype(float)
data_preprocessing['Lebar (M)'] = data_preprocessing['Lebar (M)'].str.replace(',', '').astype(float)
data_preprocessing['Tebal/Tinggi (M)'] = data_preprocessing['Tebal/Tinggi (M)'].str.replace(',', '').astype(float)

# Menampilkan data hasil akhir yang sudah diubah menjadi float
data_preprocessing[['Panjang (M)', 'Lebar (M)', 'Tebal/Tinggi (M)']]
```

# **DATA TRANSFORMATION**

```python
# Multiply 'jumlah_projek' with 'nilai_projek' and store the result in a new column 'total_nilai'
data_preprocessing['Volume Kerusakan'] = data_preprocessing['Panjang (M)'] * data_preprocessing['Lebar (M)' ]* data_preprocessing['Tebal/Tinggi (M)']

# Print the updated DataFrame
data_preprocessing
```

```python
def sub_kategori_kerusakan(Kategori):
    if Kategori in ['111 (Lubang)', '112 (Keriting)', '114 (Amblas)','116 (Kerusakan Tepi)', '117 (Retak Buaya)', '118 (Retak Garis)', '119 (Kegemukan Aspal)','120 (Terkelupas)', '160 (Permukaan Aspal Kasar)','232 (Alur)','252 (Kehilangan Permukaan)','612 (Tergerus)','710 (Longsor)']:
        return 'Ada Kerusakan'
    if Kategori in ['115 (Jembul)', '253 (Rumput Panjang)', '412 (Kerusakan Saluran)','413 (Saluran Ditumbuhi Tanaman)','431 (Pendangkalan)','471 (Tersumbat)','511 (Patok KM Rusak)','531 (Marka Jalan Pudar)','640 (Pas. Batu Amblas)','641 (Kehilangan Batuan)','840 (Kerusakan Sayap Gorong-gorong)','Kondisi Baik']:
        return 'Tidak Ada Kerusakan'
    return None  # Atau dapat diisi dengan nilai default lainnya

# Terapkan fungsi pada kolom 'Jenis Kelamin'
data_preprocessing['Kategori Kerusakan'] = data_preprocessing['Sub Kategori'].apply(sub_kategori_kerusakan)

# Tampilkan DataFrame hasil
data_preprocessing
```

```python
# Fungsi untuk menentukan kualitas berdasarkan rentang
def tentukan_volume(volume):
    if 0 <= volume <= 0.97825:
        return 'Volume Tidak Rusak'
    elif 1 <= volume <= 675:
        return 'Volume Rusak'
    else:
        return 'Volume Tidak Valid'

# Terapkan fungsi pada kolom 'Kualitas'
data_preprocessing['Kategori Volume Kerusakan'] = data_preprocessing['Volume Kerusakan'].apply(tentukan_volume)

# Tampilkan DataFrame hasil
data_preprocessing
```

```python
# Menemukan baris yang memiliki nilai NaN
rows_with_nan = data_preprocessing[data_preprocessing.isnull().any(axis=1)]

# Menampilkan baris yang memiliki nilai NaN
rows_with_nan
```

```python
# Print contoh data kosong dalam bentuk tabel
datas_preprocessing = data_preprocessing.dropna()
datas_preprocessing
```

# **FEARTURE SELECTION CHI SQUARE**

```python
from sklearn import feature_selection
# Baca file GAPED.csv dari Google Drive
path = '/content/drive/My Drive/Colab Notebooks/feature selection-Chi-Square.csv'
feature_selection = pd.read_csv(path)

#fill null values
for col in feature_selection.columns:
  feature_selection[col] = feature_selection[col].fillna(feature_selection[col].mode()[0])

# Menampilkan isi dataframe
feature_selection
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder
for col in feature_selection.columns:
  le = LabelEncoder()
  feature_selection[col] = le.fit_transform(feature_selection[col])

feature_selection[:5]
```

```python
from sklearn.feature_selection import  chi2
X = feature_selection.drop(columns=['Label'], axis=1)
y = feature_selection['Label']
```

```python
chi_scores = chi2(X, y)
```

```python
chi_scores
```

```python
chi_scores[0]
```

```python
chi_scores[1]
```

```python
# higher the chi value, higher the importance
chi_values = pd.Series(chi_scores[0], index=X.columns)
chi_values.sort_values(ascending=False, inplace=True)
chi_values.plot.bar()
```

```python
# if p-value > 0.5, lower the importance
p_values = pd.Series(chi_scores[1], index=X.columns)
p_values.sort_values(ascending=False, inplace=True)
p_values.plot.bar()
```
